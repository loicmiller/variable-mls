```python
import pandas as pd
import json
```

```python
# Source: https://github.com/dionyziz/popow/blob/9a75ab28deb5f2165856977ba9a1fce6d1301b6f/experiment-electrum/parse_blockchain.py#L17-L26
def bits_to_target(bits):
    bits = int(bits, 16)
    bitsN = (bits >> 24) & 0xff
    bitsBase = bits & 0xffffff
    target = bitsBase << (8 * (bitsN - 3))
    return int(target)
```

```python
with open("../tools/bitcoin-headers/headers.json") as f:
    headers = json.load(f)
df = pd.DataFrame(headers)
df
```

```python
df['target'] = df['bits'].apply(bits_to_target)
df
```

```python
increasing_target_indices = []
for i in range(1, len(df)):
    if df['target'][i] > df['target'][i - 1]:
        increasing_target_indices.append(i)

# Display the indices
print(f"{len(increasing_target_indices)=}")
print(increasing_target_indices)
```

```python
decreasing_diff_indices = []
for i in range(1, len(df)):
    if df['difficulty'][i] < df['difficulty'][i - 1]:
        decreasing_diff_indices.append(i)

# Display the indices
print(f"{len(decreasing_diff_indices)=}")
print(decreasing_diff_indices)
```
